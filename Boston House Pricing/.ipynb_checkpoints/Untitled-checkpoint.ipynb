```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = df.iloc[:,0:13]
y = df.iloc[:,-1]
```

```python
scaler.fit(x)
x_transformed = scaler.transform(x)
x = x_transformed
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
xgb_model = xgb.XGBRegressor(objective="reg:linear",random_state = 42)
score = cross_val_score(xgb_model,x,y,scoring="neg_mean_squared_error",cv = 10)
print(-1*score)
```

```python
from sklearn.metrics import mean_squared_error
xgb_model.fit(x_train,y_train)
```

```python
y_pred = xgb_model.predict(x_test)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
xgb_model1 = xgb.XGBRegressor()

params = {
    "colsample_bytree": (0.7, 0.3),
    "gamma": (0, 0.5),
    "learning_rate": (0.03, 0.3), # default 0.1 
    "max_depth": (2, 6), # default 3
    "n_estimators": (100, 150), # default 100
    "subsample": (0.6, 0.4)
}

search = RandomizedSearchCV(xgb_model1, param_distributions=params, random_state=42, n_iter=200, cv=3, verbose=1, n_jobs=1, return_train_score=True)

search.fit(x, y)
```

```python
search.cv_results_
```
